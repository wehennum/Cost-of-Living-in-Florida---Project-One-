```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Interest Rate': [4.045, 4.01, 3.904, 3.9675, 3.88, 6.714, 6.84, 7.072, 7.2, 7.62],
}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Principal amount (loan amount)
P = 400000

# Loan term in years (assuming a 30-year loan)
loan_term_years = 30

# Calculate the monthly payments for each interest rate
df['Monthly Payment'] = df['Interest Rate'] / 12 / 100 * P / (1 - (1 + df['Interest Rate'] / 12 / 100) ** (-loan_term_years * 12))

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Interest Rate'], df['Monthly Payment'], marker='o', linestyle='-')
plt.xlabel('Interest Rate')
plt.ylabel('Monthly Payment')
plt.title('Monthly Mortgage Payments at Different Interest Rates')
plt.grid(True, linestyle='--', alpha=0.7)

# Show the chart
plt.tight_layout()
plt.show()

```

```python
import pandas as pd

# Assuming you have a DataFrame df with 'Interest Rate' and 'Monthly Payment' columns

# Calculate the Pearson correlation coefficient (Pearson's r)
correlation = df['Interest Rate'].corr(df['Monthly Payment'])

print(f'Pearson Correlation Coefficient (r): {correlation}')
```
