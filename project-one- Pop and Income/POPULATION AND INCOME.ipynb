```python
import pandas as pd

# Sample data
data1 = {'DATE': ['2017-01-01','2018-01-01','2019-01-01', '2020-01-01', '2021-01-01', '2022-01-01'],
        'FLPOP': [20977.089,21254.926,21492.056, 21589.602, 21828.069, 22244.823]}

# Create a DataFrame
df_pop = pd.DataFrame(data1)

# Convert the 'DATE' column to datetime
df_pop['DATE'] = pd.to_datetime(df_pop['DATE'])

# Extract the year component and create a new 'Year' column
df_pop['Year'] = df_pop['DATE'].dt.year

# Drop the original 'DATE' column if it's no longer needed
df_pop = df_pop.drop(columns=['DATE'])

# Print the updated DataFrame
print(df_pop)
```

```python
import matplotlib.pyplot as plt
# Assuming you have a DataFrame named df with 'Year' and 'FLPOP' columns
years = df_pop['Year']
population = df_pop['FLPOP']
plt.figure(figsize=(10, 6))
plt.plot(years, population, marker='o', linestyle='-')
plt.title('Population Trend Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.xticks(years)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have DataFrames df_pop and df_inc
years = df_pop['Year']
population = df_pop['FLPOP']
income = df_inc['Median Income']

# Create a figure with subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Plot the population data
ax1.plot(years, population, label='Population', marker='o', linestyle='-', color='b')
ax1.set_title('Population Trend Over the Years')
ax1.set_ylabel('Population')
ax1.grid(True)
ax1.legend()

# Plot the income data
ax2.plot(years, income, label='Median Income', marker='o', linestyle='-', color='g')
ax2.set_title('Income Trend Over the Years')
ax2.set_xlabel('Year')
ax2.set_ylabel('Median Income')
ax2.grid(True)
ax2.legend()

# Calculate basic statistics
population_mean = np.mean(population)
population_median = np.median(population)
population_std = np.std(population)

income_mean = np.mean(income)
income_median = np.median(income)
income_std = np.std(income)

# Print the statistics
print(f"Population Mean: {population_mean}")
print(f"Population Median: {population_median}")
print(f"Population Standard Deviation: {population_std}")

print(f"Income Mean: {income_mean}")
print(f"Income Median: {income_median}")
print(f"Income Standard Deviation: {income_std}")

plt.show()
```
