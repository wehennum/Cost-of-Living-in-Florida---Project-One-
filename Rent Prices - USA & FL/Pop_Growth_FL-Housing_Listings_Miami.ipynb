```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

pop_growth_path = Path("pop_growthfl.txt")
popgrowthfl_df = pd.read_csv(pop_growth_path)

#Rename columns for Pop Growth
popgrowthfl_df = popgrowthfl_df.rename(columns={'FLPOP': 'Florida_Population'})
popgrowthfl_df = popgrowthfl_df.rename(columns={'DATE': 'Date'})
popgrowthfl_df
```

```python
house_listings_path = Path("house_listings.csv")
houselistings_df = pd.read_csv(house_listings_path)

#Drop Nan row
row_to_drop = 6
houselistings_df = houselistings_df.drop(index=row_to_drop)
houselistings_df
```

```python
combined_df = pd.merge(popgrowthfl_df, houselistings_df, on = ('Date'), how = 'outer')
combined_df
```

```python
# Create a new DataFrame that combines the two DataFrames
combined_df = pd.merge(popgrowthfl_df, houselistings_df, on = ('Date'), how = 'left')

# Extract the x- and y-axes for each DataFrame
combined_x = combined_df["Date"]
combined_y = combined_df["Florida_Population"]/1000
combined_y1 = combined_df["House_Listings"]

# Create the line chart
plt.figure(figsize=(15, 5))
plt.plot(combined_x, combined_y, label=("Population Growth (Florida)"))
plt.plot(combined_x, combined_y1, label=("House Listings (Florida)"))


# Add a title, labels, and grid lines
plt.title("Population Growth (Florida) vs House Listings (Florida)")
plt.ylabel("Number")
plt.xlabel("Date (Annual)")
plt.xticks(rotation=90)
#plt.ylim(0,23000)
plt.grid(True)

# Add a legend and show the plot
plt.legend()
plt.show()
```
