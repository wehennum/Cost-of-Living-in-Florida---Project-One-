```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

rent_usa_path = Path("rent_usa.csv")
usarent_df = pd.read_csv(rent_usa_path)
usarent_df.head()
```

```python
#Rename columns for USA Data
usarent_df = usarent_df.rename(columns={'CUSR0000SEHA': 'Rent Price'})
usarent_df = usarent_df.rename(columns={'DATE': 'Date'})
usarent_df.head()
```

```python
rent_fl_path = Path("rent_florida.csv")
flrent_df = pd.read_csv(rent_fl_path)
flrent_df.head()
```

```python
#Rename columns for SoFL Data
flrent_df = flrent_df.rename(columns={'CUURA320SEHA': 'Rent Price'})
flrent_df = flrent_df.rename(columns={'DATE': 'Date'})
flrent_df.head()
```

```python
#Create USA Line Chart
x_axis = usarent_df["Date"]
y_axis = usarent_df["Rent Price"]
plt.ylabel("U.S. City Average Price (USD)")
plt.xlabel("Date (Quarterly)")
plt.title("U.S. City Average Rent Prices of Primary Residence Over Time")
plt.ylabel("Rent Price (USD)")
plt.xticks(rotation=90)
plt.plot(x_axis, y_axis)
plt.grid(True)
plt.scatter(x_axis, y_axis)
plt.ylim(250,500)
plt.figure(figsize=(15, 5))
plt.show()
```

```python
#Create SoFlo Line Chart
x_axis = flrent_df["Date"]
y_axis = flrent_df["Rent Price"]
plt.title("S. FL Rent Prices of Primary Residence Over Time")
plt.ylabel("Rent Price (USD)")
plt.xlabel("Date (Quarterly)")
plt.xticks(rotation=90)
plt.ylim(250,500)
plt.plot(x_axis, y_axis)
plt.grid(True)
plt.scatter(x_axis, y_axis)
plt.figure(figsize=(15, 5))
plt.show()
```

```python
#Create USA Bar Chart
plt.bar(x_axis, y_axis)
plt.title("US Rent Prices of Primary Residence Over Time")
plt.xlabel("Date (Quarterly)")
plt.ylabel("Rent Price (USD)")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.ylim(0,450)
plt.show()
```

```python
#Create SoFlo Bar Chart
plt.bar(x_axis, y_axis)
plt.title("S. FL Rent Prices of Primary Residence Over Time")
plt.xlabel("Date (Quarterly)")
plt.ylabel("Rent Price (USD)")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.ylim(0,450)
plt.show()
```

```python
# Create a new DataFrame that combines the two DataFrames
combined_df = usarent_df.append(flrent_df)

# Extract the x- and y-axes for each DataFrame
usarent_x = usarent_df["Date"]
usarent_y = usarent_df["Rent Price"]
flrent_x = flrent_df["Date"]
flrent_y = flrent_df["Rent Price"]

# Create the line chart
plt.figure(figsize=(15, 5))
plt.plot(usarent_x, usarent_y, label=("U.S. City Average Rent Prices"))
plt.plot(flrent_x, flrent_y, label=("S. FL Rent Prices"))


# Add a title, labels, and grid lines
plt.title("S. FL vs. U.S. City Average Rent Prices of Primary Residence Over Time")
plt.ylabel("Rent Price (USD)")
plt.xlabel("Date (Quarterly)")
plt.xticks(rotation=90)
plt.ylim(280,450)
plt.grid(True)

# Add a legend and show the plot
plt.legend()
plt.show()
```

```python
# Create a new DataFrame that combines the two DataFrames
combineddf = usarent_df.append(flrent_df)

# Extract the x- and y-axes for each DataFrame
x_axis = combineddf["Date"]
usarent_y = combineddf["Rent Price"]
flrent_y = combineddf["Rent Price"]

# Create the stacked bar chart
plt.figure(figsize=(15, 5))
plt.bar(x_axis, usarent_y, label="U.S. City Average Rent Prices")
plt.bar(x_axis, flrent_y, label="S. Flo Rent Prices", bottom=usarent_y)

# Add a title, labels, and grid lines
plt.title("S. FL vs. US Rent Prices of Primary Residence Over Time (Stacked)")
plt.ylabel("Rent Price (USD)")
plt.xlabel("Date (Quarterly)")
plt.xticks(rotation=90)
plt.ylim(0,1000)
plt.grid(True)

# Add a legend and show the plot
plt.legend()
plt.show()
```
