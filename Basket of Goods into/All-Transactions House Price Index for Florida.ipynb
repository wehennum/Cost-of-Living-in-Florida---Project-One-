```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

all_transactions_path=Path("All-Transactions House Price Index for Florida.csv")
all_transactions_df = pd.read_csv(all_transactions_path)
all_transactions_df.head()
```

```python
import matplotlib.pyplot as plt

dates = all_transactions_df['DATE']
basketofgoods = all_transactions_df['FLSTHPI_CPIAUCSL']

print(dates)
print (basketofgoods)
```

```python
all_transactions_df.describe()
```

```python
plt.figure(figsize=(15,6))
x_axis_data=dates
y_axis_data=basketofgoods
plt.plot(dates, basketofgoods, color = "g")
plt.scatter(dates, basketofgoods,color="g")
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Percentage Change in Price Index")
plt.grid(True)
plt.title ("All-Transactions Florida House Price Index")
plt.show()
```

```python
plt.figure(figsize=(14,6))
x_axis_data=dates
y_axis_data=basketofgoods
plt.bar(dates, basketofgoods,color="g")
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Percentage Change in Price Index")
plt.title ("All-Transactions House Price Index")
plt.show()
```
